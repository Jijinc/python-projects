```python
banglore_house_prices= r"C:\Users\user\Downloads\archive\bengaluru_house_prices.csv"
```

```python
Data=pd.read_csv(banglore_house_prices)
```

```python
Data.head()
Data.tail()
```

```python
print (Data.isnull().sum())
```

```python
print(Data.describe())
```

```python
Data.dropna(inplace=True)
```

```python
sns.scatterplot (Data ,x='area_type',y='price')
```

```python
Price = Data.groupby('location')['price'].max().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 8))
Price.plot(kind='bar', color='skyblue')
plt.title('Top 10 area price accross banglore')
plt.xlabel('location')
plt.ylabel('price')
plt.show()
```

```python
Price = Data.groupby('total_sqft')['price'].max().sort_values(ascending=False).head(10)
```

```python
plt.pie(Price, labels=Price.index, autopct='%1.1f%%')
plt.figure(figsize=(10, 5))
plt.title('Top 10 area price in squre feet ')
plt.show()
```

```python
Price = Data.groupby('total_sqft')['price'].min().sort_values(ascending=False).head(10)
```

```python
plt.pie(Price, labels=Price.index, autopct='%1.1f%%')
plt.figure(figsize=(10, 5))
plt.title('Top 10 area price in squre feet ')
plt.show()
```

```python
plt.figure(figsize=(12, 8))
Data[['size', 'balcony', 'bath']].head(10).plot(kind='bar', stacked=True, colormap='viridis')
plt.title('bar Chart with Confirmed, bath, and balcony')
plt.xlabel('bath')
plt.ylabel('size')
plt.show()
```

```python
Price = Data.groupby('size')['price'].max().sort_values(ascending=False).head(10)
```

```python
plt.figure(figsize=(12, 8))
Price.plot(kind='bar', color='green')
plt.title('Top 10 properties size and price')
plt.xlabel('size')
plt.ylabel('price in Lakhs')
plt.show()
```
